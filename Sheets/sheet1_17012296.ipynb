```python
import numpy as np
import math
import matplotlib.pyplot as plt
import pandas as pd
data = [[10, 60, 10, 90], [20, 50, 40, 70], [
        30, 50, 30, 40], [20, 50, 20, 60], [10, 60, 30, 10]]
data_matrix = np.array(data)
```

Question 2, a) Norms of Data Matrix

```python
norms = []
for instance in data_matrix:
    norms.append(np.linalg.norm(instance))
print(norms)
```

Q2) b)  Cosine similarity matrix

```python
similarity = np.dot(data_matrix, data_matrix.T)
square_mag = np.diag(similarity)
inv_square_mag = 1 / square_mag
inv_square_mag[np.isinf(inv_square_mag)] = 0
inv_mag = np.sqrt(inv_square_mag)
cosine = similarity * inv_mag
cosine = cosine.T * inv_mag
print(cosine)
```

Q2) C) Euclidean Distance Matrix

```python
p1 = np.sum(data_matrix**2, axis=1)[:, np.newaxis]
p2 = np.sum(data_matrix**2, axis=1)
p3 = -2 * np.dot(data_matrix, data_matrix.T)
euclidean = np.round(np.sqrt(p1+p2+p3), 2)
print(euclidean)
```

Q3) b) Plot the points

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
X = pd.DataFrame(data_matrix)
X.columns = ['a1', 'a2', 'a3', 'a4']
ax.scatter(X['a1'], X['a2'], X['a4'])
plt.show()
```

 Q3) f) Eigenvalues

```python
tempZ = [[56, -32, -12], [-32, 24, -16], [-12, -16, 744]]
Z = np.array(tempZ)
eigh = np.linalg.eig(Z)
print(eigh[1])
```
